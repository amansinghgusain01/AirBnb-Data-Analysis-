```python
# Price Distribution
# Create a figure with a custom size
plt.figure(figsize=(12, 5))

# Set the seaborn theme to darkgrid
sns.set_theme(style='darkgrid')

# Create a histogram of the 'price' column of the Airbnb_df dataframe
# using sns distplot function and specifying the color as red
sns.distplot(df['price'],color=('r'))

# Add labels to the x-axis and y-axis
plt.xlabel('Price', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Add a title to the plot
plt.title('Distribution of Airbnb Prices',fontsize=15)
```

# feature enginering


```python
# creating a new colum  for price oer bed
df["price per bed"]=df["price"]/df["beds"]
df.head()
```

```python
df.groupby("neighbourhood_group")["price per bed"].mean()
```

```python
# Set the figure size (width=8, height=6)
plt.figure(figsize=(8, 6))

# Create a bar plot showing average price by neighbourhood group and room type
sns.barplot(data=df, x="neighbourhood_group", y="price", hue="room_type")

# Set the plot title with a font size of 15
plt.title("Average Price by Neighbourhood Group and Room Type", fontsize=15)

# Label the x-axis
plt.xlabel("Neighbourhood Group", fontsize=15)

# Label the y-axis
plt.ylabel("Average Price", fontsize=15)

# Adjust layout to prevent label overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between number of reviews and price
# The points are colored by 'neighbourhood_group'
sns.scatterplot(data=df, 
                x="number_of_reviews", 
                y="price", 
                hue="neighbourhood_group")

# Set the title of the plot
plt.title("Locality and Review Dependency")
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Get the room type counts
room_type_counts = df['room_type'].value_counts()

# Set the labels and sizes for the pie chart
labels = room_type_counts.index
sizes = room_type_counts.values

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add a legend to the chart
plt.legend(title='Room Type', bbox_to_anchor=(0.8, 0, 0.5, 1), fontsize='12')

# Show the plot
plt.show()
```

```python
df.columns
```

```python

# Create a pair plot (scatterplot matrix) for selected numerical columns
# 'hue' is set to 'room_type' to color the points based on different room types
sns.pairplot(data=df, 
             vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], 
             hue='room_type')

```

```python
# Geographical Distribution of AirBnb Listings

# Set the figure size for better visibility
plt.figure(figsize=(10, 7))

# Create a scatter plot with longitude on the x-axis and latitude on the y-axis
# 'hue' is used to differentiate listings by room type (e.g., Entire home, Private room, etc.)
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')

# Set the title of the plot
plt.title("Geographical Distribution of AirBnb Listing")

# Display the plot
plt.show()
```

```python
# Get the top 10 hosts by listing count
top_hosts = df['host_name'].value_counts()[:10]

# Create a bar plot of the top 10 hosts
top_hosts.plot(kind='bar', color='peru', figsize=(18, 7))

# Set the x-axis label
plt.xlabel('top10_hosts', fontsize=14)

# Set the y-axis label
plt.ylabel('total_NYC_listings', fontsize=14)

# Set the title of the plot
plt.title('top 10 hosts on the basis of no of listings in entire NYC!', fontsize=15)
```

```python
# Calculate the average price per neighbourhood group, sorted in descending order
avg_price = df.groupby("neighbourhood_group")["price"].mean().sort_values(ascending=False)

# Set the figure size for the plot
plt.figure(figsize=(10, 6))

# Create a bar plot with neighbourhood groups on x-axis and average prices on y-axis
# Use the "viridis" color palette for better visual appeal
sns.barplot(x=avg_price.index, y=avg_price.values, palette="viridis")

# Add title and axis labels
plt.title("Average Price per Neighbourhood Group")
plt.ylabel("Average Price")
plt.xlabel("Neighbourhood Group")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()

```

```python


# Step 1: Get the top 10 neighborhoods by listing count
top_10 = df['neighbourhood'].value_counts().nlargest(10).reset_index()
top_10.columns = ['neighbourhood', 'count']

# Step 2: Plot using seaborn
plt.figure(figsize=(15,6))
sns.barplot(data=top_10, x='neighbourhood', y='count', palette='pastel')

# Step 3: Customize the plot
plt.title('Listings by Top Neighborhoods in NYC', fontsize=15)
plt.xlabel('Neighbourhood', fontsize=14)
plt.ylabel('Total Listing Counts', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Set the size of the plot (width=10, height=3)
plt.figure(figsize=(10, 3))

# Create a scatter plot showing the relationship between number of reviews and price
# Hue differentiates the points based on 'room_type', and alpha controls the transparency
sns.scatterplot(data=df, x="number_of_reviews", y="price", hue="room_type", alpha=0.6)

# Add a title to the plot
plt.title("Price vs Number of Reviews by Room Type")

# Use logarithmic scale on the y-axis to handle skewness in price distribution
plt.yscale("log")

# Display the plot
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['id','host_id', 'latitude','longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count','availability_365']].corr()
corr

# plt.figure(figsize=(8, 6))

# sns.heatmap(data=corr, annot=True)
```

```python
# Calculate correlation matrix from only numeric columns
corr = df.select_dtypes(include='number').corr()
# Plot the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(data=corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, square=True)
plt.title("Correlation Heatmap of All Numeric Features")
plt.show()
```
